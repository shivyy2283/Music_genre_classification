```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled.shape
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
```

```python
pca.fit(X_scaled)
```

```python
pca_data = pca.transform(X_scaled)
pca_data.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(pca_data, y, test_size = 0.2, random_state = 0)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
rf_classifier = RandomForestClassifier(n_estimators=800,random_state=87)
rf_classifier.fit(X_train, y_train)
```

```python
y_pred = rf_classifier.predict(X_test)
print(y_pred)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
encoded_target = label_encoder.fit_transform(y_pred)
print(encoded_target)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
classifier = KNeighborsClassifier(n_neighbors = 8, metric = 'manhattan', p = 1)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
print(y_pred)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
encoded_target = label_encoder.fit_transform(y_pred)
print(encoded_target)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```
